### The simplest time series plot

```python
address = 'D:/GoogleDrive/WGU/C996-Programming in Python/jupyter_notebook/Python for Data Science Essential Training Part 1/Data/Superstore-Sales.csv'

df = pd.read_csv(address, index_col='Order Date', encoding='cp1252', parse_dates=True)
df.head()
```

```python
df['Order Quantity'].plot()
```

```python
# random_state adds seed
df2 = df.sample(n=50, random_state=25, axis=0)

plt.xlabel('Order Date')
plt.ylabel('Order Quantity')
plt.title('Superstore Sales')

df2['Order Quantity'].plot()
```
