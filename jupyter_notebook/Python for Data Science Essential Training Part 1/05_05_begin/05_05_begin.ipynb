```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sb
from pylab import rcParams

import scipy
from scipy.stats.stats import pearsonr
```

```python
%matplotlib inline
#set dimention size: data visualization 8in wide, 4in height
rcParams['figure.figsize'] = 8,4
#define plot stype
plt.style.use('seaborn-whitegrid')
```

### The Pearson Correlation

```python
address = "D:/GoogleDrive/WGU/C996-Programming in Python/jupyter_notebook/Python for Data Science Essential Training Part 1/Data/mtcars.csv"
cars = pd.read_csv(address)
#list of indexes for cars
cars.columns = ["car_names","mpg","cyl","disp","hp","drat","wt","qsec","vs","am","gear","carb"]
cars.head(3)
```

```python
sb.pairplot(cars)
```

```python
x = cars[['mpg', 'hp', 'qsec', 'wt']]
sb.pairplot(x)
```

### Using scipy to calculate the Pearson correlation coefficient

```python
mpg = cars['mpg']
hp = cars['hp']
qsec = cars['qsec']
wt = cars['wt']

pearsonr_coefficient, p_value = pearsonr(mpg,hp)
print('Pearson Correlation Coeffficient %0.3f'% (pearsonr_coefficient))
```

```python
pearsonr_coefficient, p_value = pearsonr(mpg,qsec)
print('Pearson Correlation Coeffficient %0.3f'% (pearsonr_coefficient))
```

```python
pearsonr_coefficient, p_value = pearsonr(mpg,wt)
print('Pearson Correlation Coeffficient %0.3f'% (pearsonr_coefficient))
```
